```python

import cv2
from fer import FER
import matplotlib.pyplot as plt
```

```python
import cv2
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

print("OpenCV version:", cv2.__version__)
print("TensorFlow version:", tf.__version__)
print("Matplotlib version:", matplotlib.__version__)
print("NumPy version:", np.__version__)
```

```python
#1
image_path = "face-showing-anger-microexpression-800x950.jpg"
image = cv2.imread(image_path)

# Convert the image from BGR (OpenCV format) to RGB (matplotlib format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Initialize the FER detector
detector = FER()

# Detect emotions in the image
results = detector.detect_emotions(image_rgb)

# Display the results
for result in results:
    bounding_box = result["box"]
    emotions = result["emotions"]

    # Draw the bounding box
    x, y, w, h = bounding_box
    cv2.rectangle(image_rgb, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # Display the dominant emotion
    dominant_emotion = max(emotions, key=emotions.get)
    cv2.putText(image_rgb, dominant_emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# Display the image with the detected emotions using matplotlib
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Load an image using OpenCV
image_path = "SurpriseFace.jpg"
image = cv2.imread(image_path)

# Convert the image from BGR (OpenCV format) to RGB (FER format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Initialize the FER detector
detector = FER()

# Detect emotions in the image
results = detector.detect_emotions(image_rgb)

# Output the results in text format
for i, result in enumerate(results):
    bounding_box = result["box"]
    emotions = result["emotions"]

    # Print the face index
    print(f"Face {i + 1}:")
    
    # Print the bounding box coordinates
    print(f"Bounding Box: {bounding_box}")

    # Print the detected emotions with their corresponding scores
    for emotion, score in emotions.items():
        print(f"{emotion}: {score:.2f}")

    # Print the dominant emotion
    dominant_emotion = max(emotions, key=emotions.get)
    print(f"Dominant Emotion: {dominant_emotion}\n")

print("Emotion detection completed.")
```
